```python
import yfinance as yf
import pandas as pd
# Example: Downloading financial data for Nike (NKE)
ticker = 'NKE'
nike = yf.Ticker(ticker)

# Download financial statements
financials = nike.financials
balance_sheet = nike.balance_sheet
cash_flow = nike.cashflow

# Download stock price data
stock_data = yf.download(ticker, start="2020-01-01", end="2023-12-31")

# Displaying the financials
print(financials)
print(balance_sheet)
print(cash_flow)
print(stock_data.head())
```

```python
# Financial Ratios
financials = financials.T  
financials['Gross Margin'] = financials['Gross Profit'] / financials['Total Revenue']
financials['Operating Margin'] = financials['Operating Income'] / financials['Total Revenue']
financials['Net Margin'] = financials['Net Income'] / financials['Total Revenue']

# Displaying the calculated ratios
print(financials[['Gross Margin', 'Operating Margin', 'Net Margin']])
```

```python
import matplotlib.pyplot as plt

# Plotting the Adjusted Close price over time
plt.figure(figsize=(10, 6))
plt.plot(stock_data.index, stock_data['Adj Close'])
plt.title('Nike Adjusted Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.grid(True)
plt.show()
```

```python

# Plotting the Adjusted Close price over time
plt.figure(figsize=(10, 6))
plt.plot(financials.index, financials['Total Revenue'])
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```

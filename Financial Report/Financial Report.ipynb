## Help your hedge fund manager!

This notebook is to show the creation of a financial report. we will be finding the leverage and profitability ratios of three different industries, which industry has the lowest leverage ratio, the highest profitability ratio, and what the relationship is between the leverage and profitibility ratios.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
Balance=pd.read_excel("Balance_Sheet.xlsx")
Income=pd.read_excel("Income_Statement.xlsx")
Balance
```

```python
#merge tables
df_ratios=pd.merge(Balance,Income, on=["company","company"])
```

```python
#Compute gross margin ratio
df_ratios["profitability_ratio"]= (df_ratios["Total Revenue"]-df_ratios["Total Operating Expenses"])/df_ratios["Total Revenue"]
```

```python
#computing debt-to equity ratio

df_ratios["leverage_ratio"]=df_ratios["Total Liab"]/df_ratios["Total Stockholder Equity"]
```

```python
df_ratios
```

```python
#Company with lowest average profitability ratio

print(df_ratios.pivot_table(index="comp_type_x", values="profitability_ratio"))
lowest_profitability = "fmcg"
```

```python
#Company with highest average leverage ratio
print(df_ratios.pivot_table(index="comp_type_x", values="leverage_ratio"))
highest_leverage= "real_est"
```

```python
df_real_est=df_ratios.loc[df_ratios["comp_type_x"]=="real_est"]
plot=sns.regplot(data=df_real_est, x="leverage_ratio", y="profitability_ratio")
relationship="positive"
```
